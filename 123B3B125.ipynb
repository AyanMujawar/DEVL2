```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset from seaborn's inbuilt datasets
iris = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
display(iris.head())

# Display the summary of the dataset
print("\nSummary Information:")
display(iris.info())

# Display summary statistics
print("\nSummary Statistics:")
display(iris.describe())
```

### Visualization with Matplotlib

```python
# Histogram for each numerical feature in the dataset
iris.hist(figsize=(10, 6), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Histogram of Numeric Features")
plt.show()
```

```python
# Scatter plot with Matplotlib
species_colors = {'setosa': 'blue', 'versicolor': 'green', 'virginica': 'red'}

plt.figure(figsize=(8, 6))
for species, color in species_colors.items():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species, color=color)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```

```python
# Box plot for each numerical feature
iris.plot(kind='box', figsize=(10, 6), color='blue')
plt.title("Box Plot of Numeric Features")
plt.show()
```

### Visualization with Seaborn

```python
# Pair plot for the Iris dataset with species as hue
sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle("Pair Plot of Features", y=1.02)
plt.show()
```

```python
corr_matrix = iris.select_dtypes(include='number').corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Feature Correlations")
plt.show()
```
