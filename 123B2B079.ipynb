```python
import html5lib
from bs4 import BeautifulSoup
import pandas as pd

# Open and read the local HTML file or URL
file_path = r"D:\Datasets\G.html"  # Replace with the path to your HTML file

with open(file_path, 'r') as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html5lib')

# Find all the tables in the HTML content
table_soup = soup.find_all('table')

# Initialize a list to hold DataFrames
dataframe = []

# Process each table and convert to DataFrame
for i, table in enumerate(table_soup):
    rows = table.find_all('tr')[1:]  # Skip the header row
    data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]  # Strip extra spaces
        data.append(cols)
    df = pd.DataFrame(data)
    dataframe.append(df)

# Access the 5th table (index 4)
dataframe[0]
```
