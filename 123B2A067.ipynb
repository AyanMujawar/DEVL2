```python
import pandas as pd

# Step 1: Import Data from CSV and JSON
# Assuming "data.csv" and "data.json" are in the working directory

# Load data from CSV file
csv_file = "data.csv"  # Replace with your file name
df_csv = pd.read_csv(csv_file)

# Load data from JSON file
json_file = "data.json"  # Replace with your file name
df_json = pd.read_json(json_file)

# Display the loaded DataFrames
print("DataFrame from CSV:\n", df_csv.head())  # Display first 5 rows
print("\nDataFrame from JSON:\n", df_json.head())  # Display first 5 rows

# Step 2: Data Exploration
# Shape of the data
print("\nShape of CSV DataFrame:", df_csv.shape)
print("Shape of JSON DataFrame:", df_json.shape)

# Column names
print("\nColumns in CSV DataFrame:", df_csv.columns.tolist())
print("Columns in JSON DataFrame:", df_json.columns.tolist())

# Basic statistics
print("\nBasic statistics of CSV DataFrame:\n", df_csv.describe())
print("\nBasic statistics of JSON DataFrame:\n", df_json.describe())

# Step 3: Data Manipulation
# Adding a new column to CSV DataFrame
df_csv['New_Column'] = df_csv['Existing_Column'] * 2  # Replace 'Existing_Column'

# Filtering rows where a condition is met
filtered_df = df_csv[df_csv['Existing_Column'] > 50]  # Replace 'Existing_Column'

# Renaming columns in JSON DataFrame
df_json.rename(columns={'Old_Column': 'New_Column'}, inplace=True)  # Replace 'Old_Column'

# Merging CSV and JSON DataFrames (inner join based on a common column)
merged_df = pd.merge(df_csv, df_json, on='Common_Column')  # Replace 'Common_Column'

# Step 4: Save Modified DataFrames
df_csv.to_csv("modified_data.csv", index=False)
df_json.to_json("modified_data.json", orient="records")

# Display final modified DataFrames
print("\nModified CSV DataFrame:\n", df_csv.head())
print("\nFiltered DataFrame:\n", filtered_df.head())
print("\nMerged DataFrame:\n", merged_df.head())
```
